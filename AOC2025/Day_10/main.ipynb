```python
from scipy.optimize import linprog

with open("input.txt") as f:
    input = f.read()
machines = []
result = 0
for line in input.split("\n"):
    spl = line.split(" ")
    transforms = spl[1:len(spl) - 1]
    joltages = spl[-1]
    transforms = [[int(y) for y in x[1:len(x) - 1].split(",")] for x in transforms]
    joltages = [int(x) for x in joltages[1:len(joltages) - 1].split(",")]
    equations = []
    for i in range(len(joltages)):
        equations.append([])
        for k in range(len(transforms)):
            if i in transforms[k]:
                equations[i].append(1)
            else:
                equations[i].append(0)
    res = linprog([1 for _ in range(len(transforms))], A_eq=equations, b_eq=joltages, integrality=[1 for _ in range(len(transforms))], bounds=[(0, None) for _ in range(len(transforms))])
    result += res.fun
    
print(f"Answer: {int(result)}")
```
